```python
# Humanクラスの定義
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # 年齢から大人かどうかを判定し標準出力するメソッド
    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name} さん（{self.age}歳）は大人です。")
        else:
            print(f"{self.name} さん（{self.age}歳）は大人ではありません。")


# Humanインスタンスを複数作成してリストに追加
human_list = [
    Human("山田太郎", 25),
    Human("佐藤花子", 19),
    Human("鈴木一郎", 32)
]

# リストの要素数分だけ check_adult を呼び出す
for human in human_list:
    human.check_adult()
```
